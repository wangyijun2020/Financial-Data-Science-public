```python
import numpy as np
import math
```

```python
γ1= np.random.random()
γ1
```

```python
W=np.arange(1,1000000, 1)
```

```python
U1= (W**(1-γ1)-1)/(1-γ1)
```

```python
U2= np.log(W)
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.title('Concave Utility Function')
plt.xlabel('Wealth')
plt.ylabel('Utility')
plt.plot(W, U1, linestyle='-', color='c',lw=2)
plt.plot(W,U2, linestyle='-', color='c',lw=2)
plt.show()
```

```python
#The more wealth we want to have, the higher utility we have
#Marginal utility decreases as the wealth increase
#People, who are more willing to take risks, have higher utility value
```
